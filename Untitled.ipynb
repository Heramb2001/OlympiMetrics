```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('athlete_events.csv')
region_df=pd.read_csv("noc_regions.csv")
```

```python
df=df[df['Season']=="Summer"]
df.head()
```

```python
region_df.head()
```

```python
df=df.merge(region_df,on='NOC',how='left')
```

```python
df.duplicated().any()
```

```python
df.drop_duplicates(inplace=True)#inplace=True means dataframe has to make changes permanently
```

```python
df.duplicated().any()
```

```python
df['Medal'].value_counts()
```

```python
df=pd.concat([df,pd.get_dummies(df['Medal'])],axis=1)
```

```python
df.head()
```

```python
df=df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```

```python
df[(df['NOC']=='IND') & (df['Medal']=='Gold')] #Here we will get to know the issue about dataset that team medals are taken as a individual medals
```

```python
df[df['NOC']=='IND']#The actual GOLD count of INDIA is 9 and silver is 7
```

```python
medal_tally=df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally[(medal_tally['NOC']=='IND') & (df['Medal']=='Gold')]
```

```python
medal_tally=medal_tally.groupby('region').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```

```python
medal_tally[medal_tally['NOC']=='IND']#Now the medal count is accurate
```
